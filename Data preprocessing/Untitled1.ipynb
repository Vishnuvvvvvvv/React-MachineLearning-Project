```python
recovery_thresholds={
    'flop':80,
    'Average':100,
    'hit':120,
    'Superhit':130,
    'blockbuster':150
}

def assign_category(row):
    for category,threshold in recovery_thresholds.items():
        if row['recovery-rate']<threshold:
            return category
        
    return 'All-time-Blockbuster'

ndff['category'] = ndff.apply(assign_category,axis=1)
```
