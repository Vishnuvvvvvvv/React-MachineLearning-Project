```python
top1 = merged.groupby('Director')['recovery_rate'].max().loc[lambda x: x >= 1300]
top1dir = top1.index.tolist()
top1dir[0:10]
top1 
```

```python
#To get name bw 2 codnitions
top2 = merged.groupby('Director')['recovery_rate'].max().loc[(merged.groupby('Director')['recovery_rate'].max() >= 900) & (merged.groupby('Director')['recovery_rate'].max() < 1300)]
top2dir = top2.index.tolist()
top2dir[0:10]
top2
```

```python
#To get name bw 2 codnitions
top3 = merged.groupby('Director')['recovery_rate'].max().loc[(merged.groupby('Director')['recovery_rate'].max() >= 500) & (merged.groupby('Director')['recovery_rate'].max() < 900)]
top3dir = top3.index.tolist()
top3dir[0:10]
top3 
```

```python
#To get name bw 2 codnitions
top4 = merged.groupby('Director')['recovery_rate'].max().loc[(merged.groupby('Director')['recovery_rate'].max() >= 250) & (merged.groupby('Director')['recovery_rate'].max() < 500)]
top4dir = top4.index.tolist()
top4dir[0:10]
top4 
```

```python
#To get name bw 2 codnitions
top5 = merged.groupby('Director')['recovery_rate'].max().loc[(merged.groupby('Director')['recovery_rate'].max() >= 100) & (merged.groupby('Director')['recovery_rate'].max() < 250)]
top5dir = top5.index.tolist()
top5dir[0:10]
top5 
```

```python
other  = merged.groupby('Director')['recovery_rate'].max().loc[lambda x:x<100]
otherdir =other.index.tolist()
other
```

```python
def convertDir(row):
    if row in top1dir:
        return 6
    elif row in top2dir:
        return 5
    elif row in top3dir:
        return 4
    elif row in top4dir:
        return 3
    elif row in top5dir:
        return 2
    elif row in other:
        return 1
    else:
        return 0

merged['latestDir'] = merged['Director'].apply(convertDir)
```

```python
x = newmodel.drop(columns = ['revenue','no:femCast','release_year','no:maleCrew','mainGenre'])
```

```python
y= newmodel[['revenue']]
```

```python
x.head()
```

```python
y.head()
```

```python
#Test training and MachineLearning
```

```python
newx = pd.get_dummies(x,columns=['prod_company'])
newx.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error

# Split the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(newx, y, test_size=0.2,random_state=5)

# Create and fit the Linear Regression model
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
linear_predictions = linear_model.predict(x_test)
linear_r2 = r2_score(y_test, linear_predictions)
linear_mae = mean_absolute_error(y_test, linear_predictions)

# Create and fit the Decision Tree Regression model
tree_model = DecisionTreeRegressor()
tree_model.fit(x_train, y_train)
tree_predictions = tree_model.predict(x_test)
tree_r2 = r2_score(y_test, tree_predictions)
tree_mae = mean_absolute_error(y_test, tree_predictions)

# Create and fit the Random Forest Regression model
forest_model = RandomForestRegressor(n_estimators=250)
forest_model.fit(x_train, y_train)
forest_predictions = forest_model.predict(x_test)
forest_r2 = r2_score(y_test, forest_predictions)
forest_mae = mean_absolute_error(y_test, forest_predictions)

# Print the evaluation metrics for each model
print("Linear Regression:")
print("R-squared score:", linear_r2)
print("Mean Absolute Error:", linear_mae)
print()

print("Decision Tree Regression:")
print("R-squared score:", tree_r2)
print("Mean Absolute Error:", tree_mae)
print()

print("Random Forest Regression:")
print("R-squared score:", forest_r2)
print("Mean Absolute Error:", forest_mae)
print()
```
